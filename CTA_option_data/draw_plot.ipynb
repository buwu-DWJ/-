```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly
import pandas as pd
from pylab import mpl
import datetime

mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False
und_list = ['DCE.m', 'DCE.i', 'CZCE.TA', 'CZCE.MA']
```

```python
for csd_und in und_list:
    df = pd.read_excel(f'hotmonth_position_{csd_und}.xlsx', index_col=0)
    li = list(df.index)
    new_li = list(set(li))
    new_li.sort(key=li.index)
    new_li_datetime = [datetime.datetime.strptime(str(a), "%Y%m%d") for a in new_li]
    month_position_0 = []
    month_position_1 = []
    month_position_2 = []
    for date in new_li:
        csd_df = df[df.index == date]
        month_position_0 += [int(str(csd_df['month'].values[0])[-2:])]
        month_position_1 += [int(str(csd_df['month'].values[1])[-2:])]
        month_position_2 += [int(str(csd_df['month'].values[2])[-2:])]

    df = pd.read_excel(f'hotmonth_volume_{csd_und}.xlsx', index_col=0)
    li = list(df.index)
    new_li = list(set(li))
    new_li.sort(key=li.index)
    new_li_datetime = [datetime.datetime.strptime(
        str(a), "%Y%m%d") for a in new_li]
    month_volume_0 = []
    month_volume_1 = []
    month_volume_2 = []
    for date in new_li:
        csd_df = df[df.index == date]
        month_volume_0 += [int(str(csd_df['month'].values[0])[-2:])]
        month_volume_1 += [int(str(csd_df['month'].values[1])[-2:])]
        month_volume_2 += [int(str(csd_df['month'].values[2])[-2:])]
    fig = make_subplots(
        rows=2,
        cols=1,
        subplot_titles=['按照持仓量进行排名', '按照成交量进行排名'],
        vertical_spacing = 0.08
    )
    trace_0 = go.Scatter(
        x=new_li_datetime,
        y=month_position_0,
        name='持仓量最高',
        mode='markers'
    )
    trace_1 = go.Scatter(
        x=new_li_datetime,
        y=month_position_1,
        name='持仓量第二',
        mode='markers'
    )
    trace_2 = go.Scatter(
        x=new_li_datetime,
        y=month_position_2,
        name='持仓量第三',
        mode='markers'
    )
    fig.append_trace(trace_0, 1, 1)
    fig.append_trace(trace_1, 1, 1)
    fig.append_trace(trace_2, 1, 1)

    trace_0 = go.Scatter(
        x=new_li_datetime,
        y=month_volume_0,
        name='成交量最高',
        mode='markers'
    )
    trace_1 = go.Scatter(
        x=new_li_datetime,
        y=month_volume_1,
        name='成交量第二',
        mode='markers'
    )
    trace_2 = go.Scatter(
        x=new_li_datetime,
        y=month_volume_2,
        name='成交量第三',
        mode='markers'
    )
    fig.append_trace(trace_0, 2, 1)
    fig.append_trace(trace_1, 2, 1)
    fig.append_trace(trace_2, 2, 1)
    # fig.update_layout(height=12500,
    #                   width=1800
    #                   )
    fig.show()
    plotly.offline.plot(fig, filename=f'res_{csd_und}.html')
    a = pd.DataFrame()
    a['month_position_0'] = month_position_0
    a['month_position_1'] = month_position_1
    a['month_position_2'] = month_position_2
    a['month_volume_0'] = month_volume_0
    a['month_volume_1'] = month_volume_1
    a['month_volume_2'] = month_volume_2
    a.index = new_li
    a.to_excel(f'hotmonth_{csd_und}.xlsx')
```
